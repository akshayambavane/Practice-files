# CSV Reading

```
from io import StringIO, BytesIO
```

```
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```
type(data)
```

```
pd.read_csv(StringIO(data)) # Text I/O implementation using an in-memory buffer.The initial_value argument sets the value of object. The newline argument is like the one of TextIOWrapper's constructor.
```

```
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```
df.to_csv('test.csv') # written into test.csv file
```

```
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```
print(data)
```

```
df=pd.read_csv(StringIO(data),dtype=object)
```

```
df
```

```
df['a'][1] # returns string
```

```
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':np.float,'a':'Int64'})
```

```
df
```

```
df['a'][1] # returns int
```

```
## check the datatype
df.dtypes
```

```
## Index columns and training delimiters
```

```

data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```
pd.read_csv(StringIO(data))
```

```
pd.read_csv(StringIO(data),index_col=0) # take 1st col as a index
```

```
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```
pd.read_csv(StringIO(data))
```

```
pd.read_csv(StringIO(data),index_col=False) # remove seted index col to default
```

```
## Combining usecols and index_col
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
# returns col b &c with index_col= False
```

```
## Quoting and Escape Characters. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```
pd.read_csv(StringIO(data),escapechar='\\') # removes \\
```
